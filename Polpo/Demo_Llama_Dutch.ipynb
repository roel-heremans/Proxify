```python
import re

def count_words(text):
    # Use a simple regular expression to count words
    words = re.findall(r'\b\w+\b', text)
    return len(words)
```

```python
import time
start_time = time.time()
# Playing without the temperature
from transformers import pipeline, Conversation

def generate_with_llama_dutch(question_dict_text, base_input_text):
    start_time = time.time()
    model = "BramVanroy/Llama-2-13b-chat-dutch"
    #potential usful parameters to tweak: ,"do_sample": True, "max_lengt
    chatbot = pipeline("conversational",model=model, do_sample=True,
                       temperature=0.1,
                       repetition_penalty=1.1,
                       #max_length=2000,
                       max_new_tokens = 200,
                       forced_bos_token_id=tokenizer.lang_code_to_id["zh_CN"]
                       model_kwargs={"device_map": "auto","load_in_8bit": True})
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Loading the model: {elapsed_time} seconds")

    for q_id, q in question_dict_text.items():
        start_time = time.time()
        input = f"{q}, based on the following input text: {base_input_text}"
        conversation = Conversation(input)
        output = (chatbot(conversation))[1]['content']
        print(f"{q_id}. {q} \n-----------------------------------\n{output}\n")
        end_time = time.time()
        elapsed_time = end_time - start_time
        print(f"Answering Q{q_id:02d}: {elapsed_time} seconds, Nr generated words: {count_words(output)}\n")
```
